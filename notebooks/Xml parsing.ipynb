```python
import bs4
import requests
site1 =  "https://www.gov.si/sitemap.xml"
site2 =  "https://www.gov.si/sitemap.xml/sitemap/SilverStripe-CMS-Model-SiteTree/79"
res = requests.get(site2)
```

```python
import urllib.parse as urlparse
bs_content = bs4.BeautifulSoup(res.text, "lxml")
a = [x.text for x in bs_content.find_all("loc")]
a.append("http://localhost:8888/notebooks/Documents/Projects/Faculty/web_information_extraction/crawling_gov/notebooks/Xml%20parsing.ipynb")
a.append("https://nasodiscu.si/dedovanje#top")
for x in a:
    parsed = urlparse.urlparse(x)
    url = f"{parsed.scheme}://{parsed.netloc}{parsed.path}"
    print(url)
    
```
